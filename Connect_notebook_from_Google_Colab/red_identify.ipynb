```python
#Khai bo cc th vin cn dng
import cv2
import numpy as np
from matplotlib import pyplot as plt

# c nh
image = cv2.imread('/content/sample_data/TheWheelOfFortune.jpg')

# Chuyn i khng gian mu t RGB sang HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# nh ngha mu  trong phm vi HSV
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# To mask  ch gi li cc pixel nm trong phm vi mu 
mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
mask = cv2.bitwise_or(mask1, mask2)

# p dng mask ln nh gc
result = cv2.bitwise_and(image, image, mask=mask)

# Hin th nh gc v nh kt qu s dng matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

#Hin th kt qu
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Result')

plt.show()
```
